# Downloading the data

```python
from urllib.request import urlretrieve

breast_cancer_url = 'https://archive.ics.uci.edu/static/public/17/breast+cancer+wisconsin+diagnostic.zip'
urlretrieve(breast_cancer_url, 'breast+cancer+wisconsin+diagnostic.zip')
```

```python
import zipfile

with zipfile.ZipFile('/content/breast+cancer+wisconsin+diagnostic.zip', 'r') as zip:
  zip.extractall('/content/')
```
