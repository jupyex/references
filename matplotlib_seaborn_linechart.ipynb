````python
years = range(2019, 2025)
jupyex_height = [149.0, 151.1, 153.6, 155.2, 155.4, 155.4]
trqshy_height = [150.2, 153.3, 155.9, 156.7, 157.0, 157.1]
````

````python
plt.plot(years, jupyex_height)
plt.plot(years, trqshy_height)

plt.xlabel('Year')
plt.ylabel('Height (cm)')

plt.title('@jupyex vs @trqshy')
plt.legend(['@jupyex', '@trqshy']); # to differentiate between the two lines
````

````python
# Adding markers for the data points
plt.plot(years, jupyex_height, marker = 'o')
plt.plot(years, trqshy_height, marker = 'x')

plt.xlabel('Year')
plt.ylabel('Height (cm)')

plt.title('@jupyex vs @trqshy')
plt.legend(['@jupyex', '@trqshy']); # to differentiate between the two lines
````

You may style the lines and markers using `fmt = '[marker][line][color]'`.
For instance,
```
plt.plot(years, jupyex_height, 'x-b')
```

````python
# Adding markers for the data points
plt.plot(years, jupyex_height, 'x-b')
plt.plot(years, trqshy_height, 'o--r')

plt.xlabel('Year')
plt.ylabel('Height (cm)')

plt.title('@jupyex vs @trqshy')
plt.legend(['@jupyex', '@trqshy']); # to differentiate between the two lines
````

If you don't want points to be connected, you can choose to not specify a line style in `fmt`

````python
# Adding markers for the data points
plt.plot(years, jupyex_height, 'xb')
plt.plot(years, trqshy_height, 'or')

plt.xlabel('Year')
plt.ylabel('Height (cm)')

plt.title('@jupyex vs @trqshy')
plt.legend(['@jupyex', '@trqshy']);
````

You may use Seaborn to make your charts look more beautiful

````python
sns.set_style("whitegrid")

plt.plot(years, jupyex_height)
plt.plot(years, trqshy_height)

plt.xlabel('Year')
plt.ylabel('Height (cm)')

plt.title('@jupyex vs @trqshy')
plt.legend(['@jupyex', '@trqshy']);
````
